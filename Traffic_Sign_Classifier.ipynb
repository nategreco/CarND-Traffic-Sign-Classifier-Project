### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance. 

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
# Get image to correct shape
def prepare_image(image):
    #Work with new copy
    working_image = image.copy()

    #Resize
    if working_image.shape[1] > working_image.shape[0]:
        working_image = cv2.resize(working_image, \
                           (32, int((working_image.shape[0] / working_image.shape[1]) * 32.)))
    else:
        working_image = cv2.resize(working_image, \
                           (int((working_image.shape[1] / working_image.shape[0]) * 32.), 32))
        
    #Pad to 32x32
    working_image = cv2.copyMakeBorder(working_image, \
                                       0, \
                                       32 - working_image.shape[0], \
                                       0, \
                                       32 - working_image.shape[1], \
                                       cv2.BORDER_CONSTANT, \
                                       value=0)
    
    return working_image

# Clean up image
def process_image(image):
    #Work with new copy
    working_image = image.copy()

    #Maximize contrast
    minimum = np.min(working_image)
    maximum = np.max(working_image)
    scale =  255. / (maximum - minimum)
    #working_image[...][...] = int((working_image - minimum) * scale) => Doesn't work with 3-channel image
    for x in np.nditer(working_image, op_flags=['readwrite']):
        x[...] = int((x - minimum) * scale)
    
    #Sharpen
    weight = cv2.GaussianBlur(working_image, (5,5), 20.0, 20.0)
    working_image = cv2.addWeighted(working_image, 2, weight, -1, 0)

    return working_image

# Optional - conver to HSV
def hsv_convert(image):
    #Convert to HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    return hsv_image

# Optional - convert to Grayscale
def grayscale_convert(image):
    #Convert to HSV
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    return gray_image

# Normalize
def normalize_image(image):
    #Work with new copy
    working_image = image.copy()

    #Convert to float32
    working_image = working_image.astype('float32')

    #Normalize 
    #working_image[...] = (working_image - 128.) / 128. => Doesn't work with 3-channel image
    for x in np.nditer(working_image, op_flags=['readwrite']):
        x[...] = (x - 128.) / 128.
    
    return working_image

# Intentionally vary images for training - ToDo
def augment_image(image):
    #References - http://docs.opencv.org/trunk/da/d6e/tutorial_py_geometric_transformations.html

    #Scale
    sf_x = .20 * np.random.rand() + 1. #Limit +/- 20%
    sf_y = .20 * np.random.rand() + 1. #Limit +/- 20%
    working_image = cv2.resize(image.copy(), \
                               None, \
                               fx=sf_x, \
                               fy=sf_y + .5, \
                               interpolation = cv2.INTER_CUBIC)

    #Rotate and Skew
    center_x = int(working_image.shape[1] / 2.)
    center_y = int(working_image.shape[0] / 2.)
    angle = 20.0 * np.random.rand() - 10 #Limit +/- 10 degrees
    matrix = cv2.getRotationMatrix2D((center_x, center_y), angle, 1.0)
    working_image = cv2.warpAffine(working_image, matrix, working_image.shape[:2])

    #Shift
    shift_x = int(0.15 * working_image.shape[1] * np.random.rand()) #Limit +/- 15%
    shift_y = int(0.15 * working_image.shape[0] * np.random.rand()) #Limit +/- 15%
    matrix = np.float32([[1,0,shift_x],[0,1,shift_y]])
    working_image = cv2.warpAffine(working_image, matrix, working_image.shape[:2])
    
    #Get back to correct shape
    working_image = prepare_image(working_image.copy())
    
    return working_image

def process_set(input_set):
    #Create new set
    new_set = []
    
    #Processing sets takes time, use a progress bar
    images_pbar = tqdm(input_set, unit='images')
    for image in images_pbar:
        new_image = process_image(image.copy())
        #new_image = hsv_convert(new_image)
        if n_channels == 1:
            new_image = grayscale_convert(new_image)
        new_image = augment_image(new_image)
        new_image = normalize_image(new_image)
        new_set.append(np.array(new_image))

    return np.array(new_set)
````

### Model Architecture

# Input
Input is a normalized (-1.0 to 1.0) 32x32x1 (Grayscale) image.

### Architecture
**Layer 1: Convolutional.** The output shape will be 28x28x6.

**Activation.** ReLU activation function

**Pooling.** The output shape will be 14x14x6.

**Layer 2: Convolutional.** The output shape will be 10x10x16.

**Activation.** ReLU activation function

**Pooling.** The output shape will be 5x5x16.

**Flatten.** Flatten the output shape of the final pooling layer such that it's 1D instead of 3D.

**Layer 3: Fully Connected.** This will have 120 outputs.

**Activation.** ReLU activation function

**Layer 4: Fully Connected.** This will have 84 outputs.

**Activation.** ReLU activation function

**Layer 5: Fully Connected (Logits).** This will have 43 outputs for each sign type.

### Output
Return the result of the 2nd fully connected layer.

````python
from tensorflow.contrib.layers import flatten
import tensorflow as tf

def LeNet(x):    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # Layer 1: Convolutional. Input = 32x32xx. Output = 28x28x6. x = 3 or 1 depending on n_channels
    input_height = 32
    input_width = 32
    input_depth = n_channels
    output_height = 28
    output_width = 28
    output_depth = 6
    step_width = 5
    step_height = 5
    layer_1_weights = tf.Variable(tf.truncated_normal(shape=(step_width, step_height, input_depth, output_depth), \
                                                      mean = mu, \
                                                      stddev = sigma))
    layer_1_biases = tf.Variable(tf.zeros(output_depth))
    layer_1_strides = [1, 1, 1, 1]
    layer_1_padding = 'VALID'
    layer_1 = tf.nn.conv2d(x, layer_1_weights, layer_1_strides, layer_1_padding) + layer_1_biases

    # Activation.
    layer_1 = tf.nn.relu(layer_1) #tf.nn.tanh(layer_1) #tf.nn.sigmoid(layer_1)

    # Pooling. Input = 28x28x6. Output = 14x14x6.
    input_height = output_height
    input_width = output_width
    input_depth = output_depth
    output_height = 14
    output_width = 14
    output_depth = input_depth  
    pooling_1_strides = [1, 2, 2, 1]
    pooling_1_ksize = [1, 2, 2, 1]
    pooling_1_padding = 'VALID'
    pooling_1 = tf.nn.max_pool(layer_1, pooling_1_ksize, pooling_1_strides, pooling_1_padding)

    # Layer 2: Convolutional. Output = 10x10x16.
    input_height = output_height
    input_width = output_width
    input_depth = output_depth
    output_height = 10
    output_width = 10
    output_depth = 16
    step_width = 5
    step_height = 5
    layer_2_weights = tf.Variable(tf.truncated_normal(shape=(step_width, step_height, input_depth, output_depth), \
                                                      mean = mu, \
                                                      stddev = sigma))
    layer_2_biases = tf.Variable(tf.zeros(output_depth))
    layer_2_strides = [1, 1, 1, 1]
    layer_2_padding = 'VALID'
    layer_2 = tf.nn.conv2d(pooling_1, layer_2_weights, layer_2_strides, layer_2_padding) + layer_2_biases
    
    # Activation.
    layer_2 = tf.nn.relu(layer_2)#tf.nn.tanh(layer_2) #tf.nn.sigmoid(layer_2) 
    
    # Pooling. Input = 10x10x16. Output = 5x5x16.
    input_height = output_height
    input_width = output_width
    input_depth = output_depth
    output_height = 5
    output_width = 5
    output_depth = input_depth  
    pooling_2_strides = [1, 2, 2, 1]
    pooling_2_ksize = [1, 2, 2, 1]
    pooling_2_padding = 'VALID'
    pooling_2 = tf.nn.max_pool(layer_2, pooling_2_ksize, pooling_2_strides, pooling_2_padding)

    # Flatten. Input = 5x5x16. Output = 400.
    input_height = output_height
    input_width = output_width
    input_depth = output_depth
    output_width = 400 
    output_depth = 1  
    flatten = tf.contrib.layers.flatten(pooling_2)
    
    # Layer 3: Fully Connected. Input = 400. Output = 120. 
    input_width = output_width 
    output_width = 120 
    layer_3_weights = tf.Variable(tf.truncated_normal(shape=(input_width, output_width), \
                                                      mean = mu, \
                                                      stddev = sigma))
    layer_3_biases = tf.Variable(tf.zeros(output_width))
    layer_3 = tf.matmul(flatten, layer_3_weights) + layer_3_biases
    
    # Activation.
    layer_3 = tf.nn.relu(layer_3) #tf.nn.sigmoid(layer_3) #tf.nn.tanh(layer_3)

    # Dropout.
    layer_3 = tf.nn.dropout(layer_3, dropout)
    
    # Layer 4: Fully Connected. Input = 120. Output = 84.
    input_width = output_width 
    output_width = 84 
    layer_4_weights = tf.Variable(tf.truncated_normal(shape=(input_width, output_width), \
                                                      mean = mu, \
                                                      stddev = sigma))
    layer_4_biases = tf.Variable(tf.zeros(output_width))
    layer_4 = tf.matmul(layer_3, layer_4_weights) + layer_4_biases
    
    # Activation.
    layer_4 = tf.nn.relu(layer_4) #tf.nn.sigmoid(layer_4) #tf.nn.tanh(layer_4)

    # Layer 5: Fully Connected. Input = 84. Output = 10.
    input_width = output_width 
    output_width = n_classes 
    layer_5_weights = tf.Variable(tf.truncated_normal(shape=(input_width, output_width), \
                                                      mean = mu, \
                                                      stddev = sigma))
    layer_5_biases = tf.Variable(tf.zeros(output_width))
    logits = tf.matmul(layer_4, layer_5_weights) + layer_5_biases
    
    return logits
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.
x = tf.placeholder(tf.float32, (None, 32, 32, n_channels))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, n_classes)
dropout = tf.placeholder(tf.float32)
batch = tf.Variable(0, dtype=tf.float32)

EPOCHS = 50
BATCH_SIZE = 128
dropout_rate = 0.5

#Implement decaying learning rate
start_rate = 0.002
decay_rate = 0.96
global_step = tf.Variable(0, trainable=False)
learning_rate = tf.train.exponential_decay(start_rate, \
                                           batch * BATCH_SIZE,
                                           len(X_train_processed), \
                                           decay_rate, \
                                           staircase=True)
logits = LeNet(x)
prediction = tf.nn.softmax(logits=logits) #Not used in training, for later evaluation
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, batch: 0, dropout: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    tf.set_random_seed(1)
    num_examples = len(X_train_processed)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train_processed, y_train = shuffle(X_train_processed, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train_processed[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, batch: offset, dropout: dropout_rate})
            
        validation_accuracy = evaluate(X_valid_processed, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        if validation_accuracy > 0.97:
            print("Target accuracy reached!")
            break
        
    saver.save(sess, './lenet')
    print("Model saved")
````
